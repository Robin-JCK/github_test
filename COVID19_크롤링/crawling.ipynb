```python
from bs4 import BeautifulSoup
import urllib.request
target_url = 'https://bucheon.go.kr/site/main/corona'
html = urllib.request.urlopen(target_url).read()
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
table = soup.find('table', {'class': 'table-style t-col line tbl_big'})
print(table)
```

```python
table = soup.find('table', {'class': 'table-style t-col line tbl_big'})
trs = table.find_all('tr')
for idx, tr in enumerate(trs):
  if idx > 0:
    tds = tr.find_all('td')
    city = tds[0].text.strip() # 앞뒤 여백이 있어 strip()을 사용했다.
    address = tds[3].text.strip()
    date = tds[4].text.strip()
    print(city, address, date)
```
