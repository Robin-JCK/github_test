```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import urllib.request
target_url = 'https://www.seocho.go.kr/html/notice/main.jsp#con16'
html = urllib.request.urlopen(target_url).read()
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 현재 페이지에서 table 태그 모두 선택하기
tables = soup.select('table')

# 하나의 테이블 태그 선택하기
table = tables[0]
print(table)
```

```python
# 테이블 html 정보를 문자열로 변경하기
table_html = str(table)

# 판다스의 read_html 로 테이블 정보 읽기
table_df_list = pd.read_html(table_html)

# 데이터프레임 선택하기
table_df = table_df_list[0]
```

```python
table_df['time'] = table_df.time.apply(lambda x: x.split('일 ~')[-1])
table_df['time'] = table_df.time.apply(lambda x: x.split('일~')[-1])
table_df['time'] = table_df.time.apply(lambda x: x.split('2021년 ')[-1])
table_df['time'] = table_df.time.apply(lambda x: x.split('일')[0])
table_df['time'] = table_df['time'].str.replace('월 ', '/')
table_df
```
